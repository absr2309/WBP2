```python
#Import LinearRegression to create and train linear regression models
from sklearn.linear_model import LinearRegression

# Define CSS styles for larger, purple text
styles = """
<style>
.custom-text {
    font-size: 24px; /* Adjust size as needed */
    color: purple;
}
</style>
"""

# Apply the styles to your text
text = "Ranking NHS Boards by drug misuse deaths by growth rate"
html_text = f"<span class='custom-text'>{text}</span>"

# Display the styled text
display(HTML(styles + html_text))
print("\n")

# Load dataset
df = pd.read_csv("modified_20102023.csv")

# Convert year to numeric format
df["Year Numeric"] = df["Year"] - 2000  # Normalize year

# Ensure Year Numeric is truly a number
df["Year Numeric"] = pd.to_numeric(df["Year Numeric"], errors="coerce")

# Dictionary to store growth rates
growth_rates = {}

# Compute linear regression slopes for each NHS Board
for board in df.columns[1:]:  # Skip "Year"
    X = df[["Year Numeric"]]
    y = df[board]

    model = LinearRegression()
    model.fit(X, y)

    growth_rates[board] = model.coef_[0]  # Store slope (rate of change per year)

# Sort all NHS Boards by growth rate (descending)
sorted_boards = sorted(growth_rates.items(), key=lambda x: x[1], reverse=True)

# Print ranked results
print("Ranked NHS Health Boards by Growth Rate:")
for rank, (board, rate) in enumerate(sorted_boards, start=1):
    print(f"{rank}. {board}: {rate:.2f} deaths per year")
```

```python
# Define CSS styles for larger, purple text
styles = """
<style>
.custom-text {
    font-size: 24px; /* Adjust size as needed */
    color: purple;
}
</style>
"""

# Apply the styles to your text
text = "Ranking NHS Boards by drug misuse deaths by growth rate visualised in a bar chart"
html_text = f"<span class='custom-text'>{text}</span>"

# Display the styled text
display(HTML(styles + html_text))
print("\n")

# Extract board names and their corresponding growth rates
boards = [board for board, rate in sorted_boards]
growth_values = [rate for board, rate in sorted_boards]

# Create the bar chart
plt.figure(figsize=(12, 6))
plt.barh(boards, growth_values, color="purple")  # Horizontal bar chart for clarity

plt.xlabel("Growth Rate (Deaths per Year)")
plt.ylabel("NHS Health Board")
plt.title("Ranking of NHS Health Boards by Growth Rate of Drug Misuse Deaths")
plt.gca().invert_yaxis()  # Highest growth at the top
plt.show()
```

###**Predictive Modeling Using Linear Regression for drug misuse deaths for the next three years for Greater Glasgow and Clyde, Lothian and Lanarkshire**###

Linear regression is performed to predict future drug misuse deaths for Greater Glasgow and Clyde, Lothian and Lanarkshire. Future years (2024, 2025, 2026) are predicted.

Greater Glasgow and Clyde has future drug misuse death predictions of 442 (for 2024), 463 (for 2025) and 485 (for 2026) deaths.

Lothian has future drug misuse death predictions of 196 (for 2024), 205 (for 2025) and 215 (for 2026) deaths.

Lanarkshire has future drug misuse death predictions of 191 (for 2024), 201 (for 2025) and 211 (for 2026) deaths.

The slope of the regression line represents the average change in the number of drug misuse deaths estimated yearly increase or decrease in drug misuse deaths for that specific NHS Board. A positive slope indicates an increasing trend, while a negative slope indicates a decreasing trend.

For Greater Glasgow and Clyde, it is an increase of 19.61 deaths per year.
For Lothian, it is an increase of 8.91 deaths per year.
For Lanarkshire, it is an increase of 10.00 deaths per year.

The intercept represents the predicted number of drug misuse deaths in the year 2000 according to the linear model.

Greater Glasgow and Clyde - The intercept of -54.71 means that the predicted number of drug misuse deaths in Greater Glasgow and Clyde in the year 2000 is -54.71. This negative death count isn't possible in reality; this value is a result of the model's linear trend when inferring back to that year.

Lothian - The intercept of -20.93 means that the predicted number of drug misuse deaths in Lothian in the year 2000 is -20.93.

Lanarkshire - The intercept of -57.14 means that the predicted number of drug misuse deaths in Lanarkshire in the year 2000 is -57.14.

Lanarkshire appears to have the fastest rising death rates.


The R-value, also known as the correlation coefficient, measures the strength and direction of the linear relationship between the year and drug misuse deaths. Its value ranges from -1 to +1.

An R-value close to +1 indicates a strong positive linear relationship (as the year increases, drug misuse deaths tend to increase).

All three areas, Greater Glasgow and Clyde, Lothian and Lanarkshire have a high R-value of 1.00, inferring a strong relationship between the year and the number of drug misuse deaths.


The R-squared value (0.99, 1.00, 1.00) means that approximately 99% to 100% of the variation in drug misuse deaths for these areas can be explained by the year according to the model.


P-value: The P-value is used to determine the statistical significance of the slope. It helps you decide whether the observed linear relationship between the year and drug misuse deaths in your sample is likely due to a real relationship in the population or just due to random chance.

Greater Glasgow and Clyde - the P-value is 0.06
Lothian - the P-value is 0.03
Lanarkshire - the P-value is 0.02.

A small P-value (typically less than a significance level like 0.05) suggests that the slope is statistically significant, meaning there is strong evidence of a real linear relationship. It's very likely that drug misuse deaths have been increasing in a linear fashion over the observed period in the populations of Lothian and Lanarkshire.
